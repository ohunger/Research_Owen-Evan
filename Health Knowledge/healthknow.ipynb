```python
import pandas as pd
import matplotlib.pyplot as plt
# Load the data from your csv file
df = pd.read_csv('V1A.CSV')

# Select relevant columns
relevant_columns = ['V1AD02a','V1AD02b','V1AD02c','V1AD02d','V1AD02e','V1AD02f',
                    'V1AD02g','V1AD02h','V1AD02i','V1AD02j','V1AD02k']
df_relevant = df[['PublicID'] + relevant_columns]

# Display the data
print(df_relevant)
# Create a new column 'health_knowledge_average' to store the mean of each row across the selected columns
df_relevant['health_knowledge_average'] = df_relevant[relevant_columns].mean(axis=1)

# Now the 'health_knowledge_average' for each 'PublidID' is in the 'health_knowledge_average' column
print(df_relevant[['PublicID', 'health_knowledge_average']])



# Create a histogram
plt.hist(df_relevant['health_knowledge_average'], bins=30, alpha=0.5, color='b')

plt.title('Distribution of Health Knowledge Average Scores')
plt.xlabel('Health Knowledge Average Score')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```
