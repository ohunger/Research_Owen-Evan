```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import seaborn as sns
from pathlib import Path 

# Read the CSV file
df = pd.read_csv('V2A.CSV')

# Normalize the activities to lowercase to account for different spellings
df['V2AJ01a_SP'] = df['V2AJ01a_SP'].str.lower()

# Get the counts of each activity
activity_counts = df['V2AJ01a_SP'].value_counts()

# Calculate the percentages of each activity
activity_percentages = df['V2AJ01a_SP'].value_counts(normalize=True) * 100

# Print the counts and percentages of each activity
for activity in activity_counts.index:
    print(f'Activity: {activity}')
    print(f'Count: {activity_counts[activity]}')
    print(f'Percentage: {activity_percentages[activity]:.2f}%')
    print('------')
```
