```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns


# Load the CSV files
df1 = pd.read_csv('SLEEP_ACTIGRAPHY_MODIFIED.CSV')
df2 = pd.read_csv('V1F.CSV')

# Specify the columns of interest
columns_of_interest = [
    'rest_dur_avg_all.Mod', 'rest_sleeptime_avg_all.Mod',
    'sleep_dur_avg_all.Mod', 'sleep_sleeptime_avg_all.Mod',
    'sleep_Frag_avg_all.Mod', 'sleep_WASO_avg_all.Mod',
    'sleep_SE_avg_all.Mod', 'rest_sleeptime_avg_wkday.Mod'
]

# Plot histograms for each column
for column in columns_of_interest:
    # Create a figure and axis
    fig, ax = plt.subplots(figsize=(10, 5))
    
    # Filter data based on PublicID
    data = df1.loc[df1['PublicID'].isin(df2['PublicID']), ['PublicID', column]]
    
    # Plot histogram
    ax.hist(data[column], bins=30, alpha=0.5, label=column)
    
    # Set plot title and labels
    ax.set_title(f'Histogram of {column} with PublicID')
    ax.set_xlabel(column)
    ax.set_ylabel('Frequency')
    
    # Show legend and plot
    ax.legend()
    plt.show()




```
