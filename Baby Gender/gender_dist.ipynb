```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv('CBA.CSV', encoding='ISO-8859-1')  


# Map the numerical data to actual gender
gender_map = {1: "Male", 2: "Female", 3: "Ambiguous", 'D': "Don't know"}
df['CBAA02'] = df['CBAA02'].map(gender_map)

# Count the frequencies of each gender
gender_counts = df['CBAA02'].value_counts()

# Print the count
print(gender_counts)

# Plotting the data
plt.figure(figsize=(10, 6))
gender_counts.plot(kind='bar')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

```
